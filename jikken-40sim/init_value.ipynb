```python



   
        
p=torch.tensor(10)
r=torch.tensor(32)
b=torch.tensor(8/3)
Ts = torch.tensor(0.01)

       

def RK4(xold):
        xnew=torch.zeros(3)
        k1=torch.zeros(3)
        k2=torch.zeros(3)
        k3=torch.zeros(3)
        k4=torch.zeros(3)

        k1=florenz(xold)
        k2=florenz(xold+k1*Ts/2.)
        k3=florenz(xold+k2*Ts/2.)
        k4=florenz(xold+k3*Ts)
        
        xnew[0]=xold[0]+Ts/6.0*(k1[0]+2.0*k2[0]+2.0*k3[0]+k4[0])
        xnew[1]=xold[1]+Ts/6.0*(k1[1]+2.0*k2[1]+2.0*k3[1]+k4[1])
        xnew[2]=xold[2]+Ts/6.0*(k1[2]+2.0*k2[2]+2.0*k3[2]+k4[2])
        # print(RK4_time)
        return xnew


    
def florenz(x):
        dxyz=torch.zeros_like(x)
        dxyz[0] = -p * x[0] + p * x[1]
        dxyz[1] = (r - x[2]) * x[0] - x[1] 
        dxyz[2] = x[0] * x[1] - b * x[2]
        return dxyz

   




```

```python
def init(NR,T):
    init_xyz=torch.zeros(3)
    init_xyz[0]=14.4657
    init_xyz[1]=6.9598
    init_xyz[2]=44.7489
    xyz=torch.zeros(3,210001)
    xyz[:,0]=init_xyz
    for i in range(210000):
        xyz[:,i+1]=RK4(xyz[:,i])
    fig = plt.figure(figsize=[8, 8])
    ax = fig.add_subplot(projection="3d")
    # ax.scatter(*x1[0, :], c="red")
    # ax.plot(*x1.transpose())
    ax.plot(xyz[0,:],xyz[1,:],xyz[2,:])
    plt.xlabel('x',size="large")
    plt.ylabel('y',size="large")
    plt.xticks([-20,-10,0,10,20])
    plt.yticks([-20,-10,0,10,20])
    ax.zticks=ax.set_zticks([0,10,20,30,40,50])
    ax.zlabel=ax.set_zlabel('z',size="large")
    plt.savefig('no-control.png')
    plt.show()
    return xyz
```
